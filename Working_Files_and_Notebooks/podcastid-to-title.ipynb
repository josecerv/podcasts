```python
import json

mapping = {}

with open('final_podcasts.jsonl', 'r', encoding='utf-8') as f:
    for line in f:
        item = json.loads(line.strip())
        mapping[item['podcast_id']] = item['title']

# Convert the mapping to a JSON string suitable for JavaScript
mapping_json = json.dumps(mapping, ensure_ascii=False)

# Print the JavaScript variable declaration
print('var podcastIdToTitle = ' + mapping_json + ';')
```
