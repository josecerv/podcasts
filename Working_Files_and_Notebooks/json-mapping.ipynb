```python
import json

def print_podcast_mapping(input_file):
    mapping = {}
    
    with open(input_file, 'r', encoding='utf-8') as file:
        for line in file:
            try:
                data = json.loads(line)
                if 'podcast_id' in data and 'title' in data:
                    mapping[data['podcast_id']] = data['title']
            except json.JSONDecodeError:
                continue

    js_output = "// Podcast ID to Title mapping\nvar podcastIdToTitle = "
    js_output += json.dumps(mapping, ensure_ascii=False)
    js_output += ";"
    
    print(js_output)

# Use the function
input_file = 'final_podcasts.jsonl'
print_podcast_mapping(input_file)
```
